```python
# 请在这一代码框内编写在这一节你需要的程序
# 包括但不限于
# * 数据预处理的程序，对数据的清洗、处理、分割等
# * 引入模型及模型的训练过程
# * 模型的进一步优化
# * 绘制可以表达你想法的图像
# 你可以自由的使用任意数量和任意格式的代码框，但请在最终提交的报告中注意报告的整洁与通顺


# 在数据表中选取我们感兴趣的变量来搭建机器学习模型
# 以下的分析主要借鉴于github上的 Lesson 3: Basic Python for Data Analytics (Stocks Prediction)
# 地址：https://github.com/liyangbit/PyDataRoad/blob/master/projects/Stock-prediction-with-Python/Lesson%203%20%20Basic%20Python%20for%20Data%20Analytics%20(Stocks%20Prediction).ipynb
df4 = df.drop(['highestIndex','lowestIndex','tradeDate','Year_Month','CHG','CHGPct'],axis=1,inplace=False)
df4.tail()
```

```python
from sklearn import preprocessing, svm

# We want to separate 1 percent of the data to forecast
forecast_out = int(math.ceil(0.01 * len(df4)))

# Separating the label here, we want to predict the closeIndex
forecast_col = 'closeIndex'
df4['label'] = df4[forecast_col].shift(-forecast_out)
x = np.array(df4.drop(['label'], 1))

# Scale the X so that everyone can have the same distribution for linear regression
x = preprocessing.scale(x)

# Finally We want to find Data Series of late X and early X (train) for model generation and evaluation
# x_lately: last 22 values
# x: first(2287-22) values

x_lately = x[-forecast_out:]
x = x[:-forecast_out]

# Separate label and identify it as y
y = np.array(df4['label'])
y = y[:-forecast_out]

print('Dimension of x', x.shape)
print('Dimension of y', y.shape)
```

```python
df4.corr()
```

```python
# Separation of training and testing of model by cross validation train test split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 0)
```

```python
#Build the model
from sklearn.neighbors import KNeighborsRegressor

from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Linear regression
clfreg = LinearRegression(n_jobs=-1)
clfreg.fit(X_train, y_train)

# Quadratic Regression 2
clfpoly2 = make_pipeline(PolynomialFeatures(2),Ridge())
clfpoly2.fit(X_train,y_train)

# Quadratic Regression 3
clfpoly3 = make_pipeline(PolynomialFeatures(3), Ridge())
clfpoly3.fit(X_train, y_train)

# KNN Regression
clfknn = KNeighborsRegressor(n_neighbors =2)
clfknn.fit(X_train, y_train)
```

```python
# Test the model

confidencereg = clfreg.score(X_test, y_test)
confidencepoly2 = clfpoly2.score(X_test, y_test)
confidencepoly3 = clfpoly3.score(X_test, y_test)
confidenceknn = clfknn.score(X_test, y_test)

print('The linear regression confidence is ', confidencereg)
print("The quadratic regression 2 confidence is ",confidencepoly2)
print("The quadratic regression 3 confidence is ",confidencepoly3)
print("The knn regression confidence is ",confidenceknn)
```
